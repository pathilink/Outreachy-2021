### 4.2 Impute data
[Return to Table of Contents](#Table-of-Contents)

✔️ Handling missing data.

⚠️ Only the training data will be worked on from now on.

```python
# choose argument
imputer = SimpleImputer(strategy='most_frequent')

# dataframe --> np array
data = train.values

# calculate the statistic for each column
imputer.fit(data)

def impute_data(imputer, X_train, df):
    # apply imputation
    data_trans = imputer.transform(df)

    # number of missings before
    print('Missing before: {}'.format(sum(np.isnan(data).flatten())))

    # number of missings after
    print('Missing after: {}'.format(sum(np.isnan(data_trans).flatten())))

    # list columns names
    columns = X_train.columns

    # create dataframe without missings
    X_train_clean = pd.DataFrame(data_trans, columns=columns) 
    
    return X_train_clean
```

```python
# X_train imputed
train_imputed = impute_data(imputer, train, train)
train_imputed.shape
```

✔️ Assuming that the data follow a normal distribution, I chose to keep 3.5 standard deviations in order to remove extreme outliers and not throw away too much data. This way it is possible to eliminate only the really discrepant data.

```python
# select rows: z-score <= 3.5
train_imputed = train_imputed[(np.abs(stats.zscore(df_zscore)) <= 3.5).all(axis=1)].copy()

print('The data has {} rows and {} columns.'.format(train_imputed.shape[0], train_imputed.shape[1]))
```

### 4.4 Correlation
[Return to Table of Contents](#Table-of-Contents)

✔️ Calculating the correlation and using it to define columns to be removed from the dataset serves to decrease the possibility of multicollinearity. I chose to remove columns that presented a correlation greater than 0.85: if two or more variables have a high correlation with each other, they are saying the same thing.

```python
# delete high correlation columns

threshold = 0.85
corr_matrix = train_imputed.corr().abs()
# corr_matrix.head()

upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
# upper.head(10)

# select columns with correlation above threshold
to_drop = [column for column in upper.columns if any(upper[column] > threshold)]
print(f'There are {len(to_drop)} columns to drop.')

# drop columns from to_drop
train_dropped = train_imputed.drop(to_drop, axis=1) 
```

```python
print('Model variables:')
print(f'- Independent: {train_dropped.columns[:-1]}')
print(f'- Dependent: {train_dropped.columns[-1]}')
print(' ')
print('The data has {} rows and {} columns.'.format(train_dropped.shape[0], train_dropped.shape[1]))
```

### 4.5 Normalization
[Return to Table of Contents](#Table-of-Contents)

✔️ Transforms the numerical variables to the same scale. This prevents variables with large values from prevailing over smaller values in the model. In robust scaling, we scale each feature of the data set by subtracting the median and then dividing by the interquartile range - Q1(x) is the first quartile of the attribute x, Q2(x) is the median, and Q3(x) is the third quartile. This method performs well with outliers.

$$ 
x = xi - Q2(x) / Q3(x) - Q1(x) 
$$

```python
# create a scaler object
scaler = RobustScaler()

# fit and transform the data
train_normalization = pd.DataFrame(scaler.fit_transform(train_dropped), columns=train_dropped.columns)

# normalized data
train_normalization.head()
```
